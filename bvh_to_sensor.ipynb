```python
import numpy as np
import matplotlib.pyplot as plt
from sensor import extract_vir_imu
import matplotlib
import os
import pandas as pd
# from sensor_util import toMag_wrist, toMag_ring, downsample, align_time, csv_to_final

# from moviepy.video.io.VideoFileClip import VideoFileClip
%load_ext autoreload
%autoreload 2
```

```python
activities = ['walking forward', 'walking counter-clockwise', 'walking clockwise', 'climbing up stairs', 
              'climbing down stairs', 'running', 'jumping', 'sitting', 'standing', 'lying', 'riding up elevator',
             'riding down elevator']

# activities = [ 'walking counter-clockwise', 'walking clockwise', 'jumping']


# convert bvh file to sensor data 
sensor_dir = 'sensor_data/usc-had_50/ideal/ik'
bvh_resized_dir = 'bvh_resized/usc-had_50/ik'
dataset = 'usc-had'
resized_dir = 'resized'

os.makedirs(os.path.join(sensor_dir), exist_ok = True)
for act in activities:
    os.makedirs(os.path.join(sensor_dir,act), exist_ok = True)
    os.makedirs(os.path.join(sensor_dir,act, resized_dir), exist_ok = True)
    bvh_word_resized = os.path.join(bvh_resized_dir,act)
    
    bvh_files_resized = [os.path.join(bvh_word_resized, f) for f in os.listdir(bvh_word_resized) if f.endswith('.bvh')]
    
    print(act)
    print(len(bvh_files_resized))
        
    for file in bvh_files_resized:
        dirname, filename = os.path.split(file)
        filename, ext = os.path.splitext(filename)
        sensor_word_resized_dir = os.path.join(sensor_dir,act, resized_dir)
        print(f'{sensor_word_resized_dir}/{filename}_acc.npz')

        #don't process the files that has IMU data already extracted 
        if os.path.exists(f'{sensor_word_resized_dir}/{filename}_acc.npz'):
            continue
            
        try:
            extract_vir_imu(file , f'{sensor_word_resized_dir}/{filename}_acc.npz', 
                        f'{sensor_word_resized_dir}/{filename}_gyro.npz', dataset)
        except:
            print('error in IMU extraction')
    
```

```python
locations = ['chest', 'forearm', 'head', 'shin', 'thigh', 'upperarm', 'waist']
#load_data from sensor npz file, all_loc is a boolean indicating using all realworld sensor locations
def load_data(file_path, loc, dataset):
    sensor_data = np.load(file_path)
#     print(sensor_data.keys())
    sensor_loc =[]
    for key in sensor_data.keys():
        sensor_loc.append(key)
#     sensor_loc = ['Spine3', 'Left_wrist', 'Head', 'Left_ankle', 'Left_knee', 'Left_elbow', 'Spine1']
#     sensor_loc = sensor_data.keys()
    print(sensor_loc)
    min_length = len(sensor_data[sensor_loc[0]])
    for sen_loc in sensor_loc:
        if len(sensor_data[sen_loc]) <min_length:
            print(f'min length changed from {min_length} to {len(sensor_data[sen_loc])}')
            min_length = len(sensor_data[sen_loc])
            
    
    if loc == 'all_loc':
        if dataset =='realworld':
            data = sensor_data["Spine3"][:min_length] #chest
            data = np.hstack((data, sensor_data[f"Left_wrist"][:min_length]))    #forearm 
            data = np.hstack((data, sensor_data[f"Head"][:min_length]))    #head 
            data = np.hstack((data, sensor_data[f"Left_ankle"][:min_length]))    #shin 
            data = np.hstack((data, sensor_data[f"Left_knee"][:min_length]))    #thigh    could also do Left_hip 
            data = np.hstack((data, sensor_data[f"Left_elbow"][:min_length]))    #upperarm 
            data = np.hstack((data, sensor_data[f"Spine1"][:min_length]))    #waist 
            return data 
        elif dataset =='pamap':
            data = sensor_data["Right_wrist"][:min_length] #hand
            data = np.hstack((data, sensor_data[f"Spine3"][:min_length]))    #chest 
            data = np.hstack((data, sensor_data[f"Right_ankle"][:min_length]))    #ankle
            return data 
        elif dataset == 'usc-had':
            return sensor_data["Right_knee"]  
        else:
            raise ValueError(f"{dataset} is not valid ")
            
    elif loc == 'wrist':
        if dataset =='realworld':
            return sensor_data["Left_wrist"]
        elif dataset =='pamap':
            return sensor_data["Right_wrist"]  
        else:
            raise ValueError(f"{dataset} is not valid ")
    else :
        raise ValueError("loc is not wrist or all_loc ")

    
# activities = [
#     'lying',
#     'sitting',
#     'standing',
#     'walking',
#     'running',
#     'cycling',
#     'Nordic walking',
#     'Ascending stairs',
#     'Descending stairs',
#     'vacuum cleaning',
#     'Ironing',
#     'rope jumping'
# ]
activities = ['walking forward', 'walking counter-clockwise', 'walking clockwise', 'climbing up stairs', 
              'climbing down stairs', 'running', 'jumping', 'sitting', 'standing', 'lying', 'riding up elevator',
             'riding down elevator']
act_to_idx = {activities[i]: i for i in range(len(activities))}

#concate the sensor data into 1 file
sensor_dir = 'sensor_data/usc-had_50/ideal/ik'
bvh_resized_dir = 'bvh_resized/usc-had_50/ik'
data_dir = 'data/usc-had_50'
ik_mode = '_ik'
noise_mode = ''
dataset = 'usc-had'

resized_dir = 'resized'
# locations = ['all_loc', 'wrist']
locations = ['all_loc']

os.makedirs(data_dir, exist_ok = True)

for loc in locations:

    X_resized = None
    Y_resized = None

    for act in activities:
        bvh_word_resized = os.path.join(bvh_resized_dir,act)

        bvh_files_resized = [os.path.join(bvh_word_resized, f) for f in os.listdir(bvh_word_resized) if f.endswith('.bvh')]

        for file in bvh_files_resized:
            dirname, filename = os.path.split(file)
            filename, ext = os.path.splitext(filename)
            print(file)

            data_filepath = os.path.join(sensor_dir,act, resized_dir, f'{filename}_acc.npz')     
            if not os.path.exists(data_filepath):
                print('sensor data not exist')
                continue
            
            data = load_data(data_filepath, loc, dataset)
            print(data.shape)

            if X_resized is None:
                X_resized=data
                Y_resized=np.ones(data.shape[0]) *act_to_idx[act] 
            else:
                X_resized=np.concatenate((X_resized, data), axis=0)
                Y_resized=np.concatenate((Y_resized, np.ones(data.shape[0])*act_to_idx[act] ), axis=0)

    print(X_resized.shape)
    print(Y_resized.shape)
    print(np.unique(Y_resized))
    # np.savez(os.path.join(sensor_dir, 'original_all_loc_raw'), x=X_original, y=Y_original)
    np.savez(os.path.join(data_dir, f'resized_{loc}_raw{ik_mode}{noise_mode}'), x=X_resized, y=Y_resized)
```
